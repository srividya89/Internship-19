```python
from bs4 import BeautifulSoup
```

```python
import requests
```

```python
page=requests.get("https://www.imdb.com/india/top-rated-indian-movies")
page
```

```python
page.content
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
print(soup.prettify())
```

```python
names=soup.find_all('td',class_="titleColumn")
names
```

```python
Series_name=[]
for i in names:
    for j in i.find_all("a"):
        Series_name.append(j.text.replace("\n",""))
Series_name
```

```python
ratings=soup.find_all('td',class_="ratingColumn imdbRating")
ratings
```

```python
Series_rating=[]
for i in ratings:
     for j in i.find_all("strong"):
        Series_rating.append(j.text.replace("\n",""))
    
Series_rating
```

```python
years=soup.find_all('span',class_="secondaryInfo")
years                    
```

```python
Series_year=[]
for i in years:
    Series_year.append(i.text.replace("\n",""))
Series_year
```

```python
ranks=[]
for i in range(1,len(Series_name)+1):
    ranks.append(i)
ranks    
```

```python
import pandas as pd
topmovies=pd.DataFrame({})
topmovies['MOVIE rank']=ranks
topmovies['Top Indian Movie Names']=Series_name
topmovies[' IMDB rating']=Series_rating
topmovies['year of release']=Series_year

topmovies.style.hide_index()
```

```python
len(topmovies)
```

```python
from bs4 import BeautifulSoup
```

```python
import requests
```

```python
HEADERS=({'User-Agent':'Mozilla/5.0 (X11;Linux X86_64) AppleWebKit/537.36 (KHTML,like Gecko) Chrome/44.0.2403.157 Safari/537.36','Accept-Language':'en-US,en;q=0.5'})
```

```python
url="https://www.amazon.in/s?k=mobiles+under+20000&rh=n%3A1389432031&ref=nb_sb_noss"
page=requests.get(url,headers=HEADERS)
page
```

```python
page.content
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
print(soup.prettify())
```

```python
mobileph=soup.find_all('span',class_="a-size-medium a-color-base a-text-normal")
mobileph
```

```python
mobilephnames=[]
for i in mobileph:
    mobilephnames.append(i.text.replace("\n",""))
mobilephnames
```

```python
len(mobilephnames)
```

```python
prices=soup.find_all('span',class_="a-price-whole")
prices
```

```python
price=[]
for i in prices:
    price.append(i.text.replace("\n",""))
price
```

```python
price.remove(price[30])
price
```

```python
images=soup.find_all('div',class_="a-section aok-relative s-image-fixed-height")
images
```

```python
image=[]
for i in images:
    for j in i.find_all("img",class_="s-image"):
        image.append(j.get("src"))
image
```

```python
len(image)
```

```python
avgR=soup.find_all('span',class_="a-icon-alt")
avgR
```

```python
AvgR=[]
for i in avgR:
     AvgR.append(i.text.replace("\n",""))
AvgR
```

```python
AvgR.remove(AvgR[30])
AvgR.remove(AvgR[30])
AvgR.remove(AvgR[30])
AvgR
```

```python
Nos=[]
for i in range(1,len(mobilephnames)+1):
    Nos.append(i)
Nos    
```

```python
import pandas as pd
AmazonMobilephones=pd.DataFrame({})
AmazonMobilephones['No.']=Nos    
AmazonMobilephones['Mobile Phone Names']=mobilephnames
AmazonMobilephones['Price']=price
AmazonMobilephones['Image']=image
AmazonMobilephones['AverageRating']=AvgR
AmazonMobilephones.style.hide_index()
```

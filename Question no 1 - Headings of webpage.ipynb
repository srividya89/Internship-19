```python
from bs4 import BeautifulSoup
```

```python
import requests
```

```python
page=requests.get("https://en.wikipedia.org/wiki/Main_Page")
page
```

```python
page.content
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
print(soup.prettify())
```

```python
All_headings=soup.find_all('span',class_="mw-headline")
All_headings
```

```python
name=soup.find_all("span",class_="mw-headline")
name
```

```python
Series_name=[]
for i in name:
    Series_name.append(i.text)
Series_name
```

```python
Heading_names=[]
for i in name:
    Heading_names.append(i.text.replace("\xa0...",""))
Heading_names
```

```python
Nos=[]
for i in range(1,len(Heading_names)+1):
    Nos.append(i)
Nos    
```

```python
import pandas as pd
Headings=pd.DataFrame({})
Headings['No.']=Nos
Headings['Headings of webpage']=Heading_names
Headings.style.hide_index()
```

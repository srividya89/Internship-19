```python
from bs4 import BeautifulSoup
```

```python
import requests
```

```python
page=requests.get("https://bookpage.com/reviews")
page
```

```python
page.content
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
print(soup.prettify())
```

```python
bnames=soup.find_all('h4',class_="italic")
bnames
```

```python
BSeries_name=[]
for i in bnames:
    BSeries_name.append(i.text.replace("\n",""))
BSeries_name
```

```python
len(BSeries_name)
```

```python
Anames=soup.find_all('p',class_="sans bold")
Anames
```

```python
ASeries_name=[]
for i in Anames:
    ASeries_name.append(i.text.replace("\n",""))
ASeries_name
```

```python
bgenre=soup.find_all('p',class_="genre-links hidden-phone")
bgenre
```

```python
genre_name=[]
for i in bgenre:
    genre_name.append(i.text.replace("\n",""))
genre_name
```

```python
rev=soup.find_all('div',class_="flex-article-content")
rev
```

```python
review=[]
for i in rev:
  
     for j in i.find_all("p"):
         review.append(j.text.replace("\n",""))
review           
```

```python
review.remove(review[14])
review  
```

```python
review.remove(review[17])
review  
```

```python
review.remove(review[20])
review  
```

```python
reviewsforbook = [] 
for index in range(2, len(review), 3):
    
    reviewsforbook.append(review[index])
print(reviewsforbook)
```

```python
Nos=[]
for i in range(1,len(BSeries_name)+1):
    Nos.append(i)
Nos    
```

```python
import pandas as pd
BookDetails=pd.DataFrame({})
BookDetails['No.']=Nos
BookDetails['BOOK Names']=BSeries_name
BookDetails['Author Names']=ASeries_name
BookDetails['GENRE']=genre_name
BookDetails['Review']=reviewsforbook
BookDetails.style.hide_index()
```

```python
from bs4 import BeautifulSoup
```

```python
import requests
```

```python
page=requests.get("https://www.icc-cricket.com/rankings/mens/player-rankings/odi/batting")
page
```

```python
page.content
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
print(soup.prettify())
```

```python
ODIbatting_Teams=soup.find_all('table',class_="table rankings-table")
ODIbatting_Teams
```

```python
ODIbattingg_Teams=soup.find_all('td',class_="table-body__cell rankings-table__name name")
ODIbattingg_Teams
```

```python
bat_mens=[]
for i in ODIbattingg_Teams:
    for j in i.find_all("a"):
        bat_mens.append(j.text.replace("\n",""))
bat_mens
```

```python
bat_mens_teams=soup.find_all('span',class_="table-body__logo-text")
bat_mens_teams
```

```python
coun_mens=[]
for i in bat_mens_teams:
     coun_mens.append(i.text.replace("\n",""))
coun_mens
```

```python
ODIbattinggy_Teams=soup.find_all('tr',class_="rankings-block__banner")
ODIbattinggy_Teams
```

```python

batt_mens=[]
for i in ODIbattinggy_Teams:
    for j in i.find_all("td"):
        batt_mens.append(j.text.replace("\n",""))
batt_mens
```

```python
bat_mens.insert(0,batt_mens [1])
bat_mens
```

```python
coun_mens.insert(0,batt_mens [2])
coun_mens
```

```python
rating=[]
for i in ODIbatting_Teams:
    for j in i.find_all("td",class_="table-body__cell rating"):
        rating.append(j.text.replace("\n",""))
rating
```

```python
rating.insert(0,batt_mens [3])
rating
```

```python
career_bestrate=[]
for i in ODIbatting_Teams:
    for j in i.find_all("td",class_="table-body__cell u-text-right u-hide-phablet"):
        career_bestrate.append(j.text.replace("\n",""))
career_bestrate
```

```python
career_bestrate.insert(0,batt_mens[4])
career_bestrate
```

```python
Rankno=[]
for i in range(1,len(bat_mens)+1):
    Rankno.append(i)
Rankno    
```

```python
import pandas as pd
ODI_Batsmen_Team=pd.DataFrame({})
ODI_Batsmen_Team['Position']=Rankno  
ODI_Batsmen_Team[' ODI Batsmen Names']=bat_mens
ODI_Batsmen_Team['Country']=coun_mens
ODI_Batsmen_Team['Rate']=rating
ODI_Batsmen_Team['Career Best Rating']=career_bestrate
ODI_Batsmen_Team
```

```python
from bs4 import BeautifulSoup

```

```python
import requests
```

```python
page=requests.get("https://www.icc-cricket.com/rankings/mens/player-rankings/odi/bowling")
page
```

```python
page.content
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
print(soup.prettify())
```

```python
ODITeams=soup.find_all('table',class_="table rankings-table")
ODITeams
```

```python
ODITeamss=soup.find_all('tr',class_="rankings-block__banner")
ODITeamss
```

```python

bow_mens=[]
for i in ODITeamss:
    for j in i.find_all("td"):
        bow_mens.append(j.text.replace("\n",""))
bow_mens
```

```python
ODIbowlmens=soup.find_all('td',class_="table-body__cell rankings-table__name name")
ODIbowlmens
```

```python
bowl_mens=[]
for i in ODIbowlmens:
    for j in i.find_all("a"):
        bowl_mens.append(j.text.replace("\n",""))
bowl_mens
```

```python
ODIbowlcon=soup.find_all('span',class_="table-body__logo-text")
ODIbowlcon
```

```python
cons=[]
for i in ODIbowlcon:
     cons.append(i.text.replace("\n",""))
cons
```

```python
ODIbowlrate=soup.find_all('td',class_="table-body__cell rating")
ODIbowlrate
```

```python
rates=[]
for i in ODIbowlrate:
     rates.append(i.text.replace("\n",""))
rates
```

```python
ODIbowlrating=soup.find_all('td',class_="table-body__cell u-text-right u-hide-phablet")
ODIbowlrating
```

```python
rating=[]
for i in ODIbowlrating:
     rating.append(i.text.replace("\n",""))
rating
```

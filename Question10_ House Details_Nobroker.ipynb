```python
soup=BeautifulSoup(page.content)
soup
```

```python
print(soup.prettify())
```

```python
Housenames=soup.find_all('h2',class_="heading-6 font-semi-bold nb__1AShY")
Housenames
```

```python
names=[]
for i in Housenames :
    names.append(i.text.replace("\n",""))
names
```

```python
len(names)
```

```python
locations=soup.find_all('div',class_="nb__2CMjv")
locations
```

```python
locs=[]
for i in locations :
    locs.append(i.text.replace("\n",""))
locs
```

```python
Area=soup.find_all('span',id ="loc")
Area
```

```python
areas=[]
for i in Area:
    areas.append(i.text.replace("\n",""))
areas
```

```python
locarea= []
for index in range(0,10):
    locarea.append(areas[0])

print(locarea)
```

```python
len(locarea)
```

```python
Sqft=soup.find_all('div',class_="nb__3oNyC")
Sqft
```

```python
Sqfts=[]
for i in Sqft:
    Sqfts.append(i.text.replace("\n",""))
Sqfts
```

```python
Emis=soup.find_all('div',class_="font-semi-bold heading-6")
Emis
```

```python
emisnp=[]
for i in Emis:
    emisnp.append(i.text.replace("\n",""))
emisnp
```

```python
emii = []
for index in range(1, len(emisnp), 3):
     emii.append(emisnp[index])

print(emii)
```

```python
price = []
for index in range(2, len(emisnp), 3):
     price.append(emisnp[index])

print(price)
```

```python
Nos=[]
for i in range(1,len(names)+1):
    Nos.append(i)
Nos    
```

```python
import pandas as pd
HouseDetails=pd.DataFrame({})
HouseDetails['No']=Nos
HouseDetails['House Title']=names
HouseDetails['Location']=locs
HouseDetails['Area']=locarea
HouseDetails['LivingArea']=Sqfts
HouseDetails['EMI/month']=emii
HouseDetails['HousePrice']=price
HouseDetails.style.hide_index()
```

```python
from bs4 import BeautifulSoup
```

```python
import requests
```

```python
page=requests.get("https://www.icc-cricket.com/rankings/mens/team-rankings/odi")
page
```

```python
page.content
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
print(soup.prettify())
```

```python
ODITeams=soup.find_all('span',class_="u-hide-phablet")
ODITeams
```

```python
ODITeams_name=[]
for i in ODITeams:
      ODITeams_name.append(i.text.replace("\n",""))
ODITeams_name
```

```python
ODITeams_name_list = list(filter(None,ODITeams_name))
print(ODITeams_name_list)
```

```python
matches=soup.find_all('table',class_="table")
matches
```

```python
recordss=[]
for i in matches:
    for j in i.find_all('td'):
         recordss.append(j.text.replace("\n",""))
recordss
```

```python
matches_score = []
for index in range(2, len(recordss), 5):
    matches_score.append(recordss[index])
print(matches_score)
```

```python
points = [] 
for index in range(3, len(recordss), 5):
    
    points.append(recordss[index])
print(points)
```

```python
len(points)
```

```python
Rating = [] 
for index in range(4, len(recordss),5):
    
    Rating.append(recordss[index])

print(Rating )
```

```python
Rank = [] 
for index in range(0, len(recordss),5):
    
    Rank.append(recordss[index])

print(Rank)
```

```python
import pandas as pd
ODIMENMATCH=pd.DataFrame({})
ODIMENMATCH['Position']=Rank
ODIMENMATCH['ODI TEAMS Men Names']=ODITeams_name_list
ODIMENMATCH['Matches']=matches_score
ODIMENMATCH['Points']=points
ODIMENMATCH['Rating']=Rating
ODIMENMATCH.style.hide_index()
```

```python
import pandas as pd
ODIMENMATCH=pd.DataFrame({})
ODIMENMATCH['Position']=Rank[:10]
ODIMENMATCH['ODI TEAMS Men Names']=ODITeams_name_list[:10]
ODIMENMATCH['Matches']=matches_score[:10]
ODIMENMATCH['Points']=points[:10]
ODIMENMATCH['Rating']=Rating[:10]
ODIMENMATCH.style.hide_index()
```

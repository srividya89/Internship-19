```python
from bs4 import BeautifulSoup
```

```python
import requests
```

```python
page=requests.get("https://www.icc-cricket.com/rankings/womens/player-rankings/odi/batting")
page
```

```python
page.content
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
print(soup.prettify())
```

```python
ODIWBTeams=soup.find_all('tr',class_="rankings-block__banner")
ODIWBTeams
```

```python
odiBwomens=[]
for i in ODIWBTeams:
    for j in i.find_all("td"):
        odiBwomens.append(j.text.replace("\n",""))
odiBwomens
```

```python
ODIWBTeamsnames=soup.find_all('td',class_="table-body__cell rankings-table__name name")
ODIWBTeamsnames
```

```python
ODINAMES=[]
for i in ODIWBTeamsnames:
     ODINAMES.append(i.text.replace("\n",""))
ODINAMES
```

```python
ODIWBTeamscon=soup.find_all('span',class_="table-body__logo-text")
ODIWBTeamscon
```

```python
odiBcons=[]
for i in ODIWBTeamscon:
    odiBcons.append(i.text.replace("\n",""))
odiBcons
```

```python
ODIWBTeamsrate=soup.find_all('td',class_="table-body__cell rating")
ODIWBTeamsrate
```

```python
odiBrate=[]
for i in ODIWBTeamsrate:
    odiBrate.append(i.text.replace("\n",""))
odiBrate
```

```python
ODIWBTeamsCBrate=soup.find_all('td',class_="table-body__cell u-text-right u-hide-phablet")
ODIWBTeamsCBrate
```

```python
odiCBrate=[]
for i in ODIWBTeamsCBrate:
    odiCBrate.append(i.text.replace("\n",""))
odiCBrate
```

```python
ODINAMES.insert(0,odiBwomens[1])
ODINAMES
```

```python
odiBcons.insert(0,odiBwomens[2])
odiBcons
```

```python
odiBrate.insert(0,odiBwomens[3])
odiBrate
```

```python
odiCBrate.insert(0,odiBwomens[4])
odiCBrate
```

```python
Rankpos=[]
for i in range(1,len(ODINAMES)+1):
    Rankpos.append(i)
Rankpos    
```

```python
import pandas as pd
ODIwomenBTeams=pd.DataFrame({})
ODIwomenBTeams['Position']=Rankpos    
ODIwomenBTeams['ODI women TEAMS Names']=ODINAMES
ODIwomenBTeams['country']=odiBcons
ODIwomenBTeams['rating']=odiBrate
ODIwomenBTeams['careerbestRating']=odiCBrate
ODIwomenBTeams
```

```python
import pandas as pd
ODIwomenBTeams=pd.DataFrame({})
ODIwomenBTeams['Position']=Rankpos[:10]   
ODIwomenBTeams['ODI women Batting TEAMS Names']=ODINAMES[:10]
ODIwomenBTeams['country']=odiBcons[:10]
ODIwomenBTeams['rating']=odiBrate[:10]
ODIwomenBTeams['careerbestRating']=odiCBrate[:10]
ODIwomenBTeams.style.hide_index()
```

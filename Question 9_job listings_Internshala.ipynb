```python
from bs4 import BeautifulSoup
```

```python
import requests
```

```python
page=requests.get("https://internshala.com/fresher-jobs")
page
```

```python
page.content
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
print(soup.prettify())
```

```python
jobtitles=soup.find_all('div',class_="heading_4_5 profile")
jobtitles
```

```python
jobtitless=[]
for i in jobtitles:
    jobtitless.append(i.text.replace("\n",""))
jobtitless
```

```python
len(jobtitless)
```

```python
Cpname=soup.find_all('div',class_="heading_6 company_name")
Cpname
```

```python
compnames=[]
for i in Cpname:
    compnames.append(i.text.replace("\n",""))
compnames
```

```python
Ctc=soup.find_all('div',class_="item_body")
Ctc
```

```python
ctcs=[]
for i in Ctc:
    ctcs.append(i.text.replace("\n",""))
ctcs
```

```python
sal = []
for index in range(1, len(ctcs), 3):
    sal.append(ctcs[index])

print(sal)
```

```python
apply=soup.find_all('div',class_="item_body")
apply
```

```python
applyd = []
for index in range(2, len(ctcs), 3):
    applyd.append(ctcs[index])

print(applyd)
```

```python
Nos=[]
for i in range(1,len(jobtitless)+1):
    Nos.append(i)
Nos    
```

```python
import pandas as pd
Internshala=pd.DataFrame({})
Internshala['No.']=Nos
Internshala['Job Title']=jobtitless
Internshala['Company Name']=compnames
Internshala['CTC']=sal
Internshala['ApplyDate']=applyd
Internshala.style.hide_index()
```
